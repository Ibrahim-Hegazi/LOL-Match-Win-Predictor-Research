```python
def get_puuid(gameName=None, tagLine=None, region='americas'):
    '''Retrieve a player's PUUID (Player Universally Unique Identifier) from the Riot Games API.

    The PUUID can be looked up using either Riot ID (modern
    identifier consisting of game name + tag line). At least one 
    identification method must be provided.

    Args:
        gameName (str, optional): The player's Riot ID game name (the part before the #).
            Must be provided together with tagLine. This is the modern player identifier.
            Example: "NOOB HIMA 55" (for Riot ID "NOOB HIMA 55#EUNE")
        tagLine (str, optional): The player's Riot ID tag line (the part after the #).
            Must be provided together with gameName. Case-insensitive.
            Example: "EUNE" or "eune"
            DONT INCLUDE THE HASHTAG WHILE PASSING VARIABLES TO THE FUNCTION
        region (str, optional): The Riot API routing region to query.
            Determines which regional API server to hit.
            Available values:
                - 'americas'  (NA, BR, LAN, LAS)
                - 'europe'    (EUW, EUNE, TR, RU)
                - 'asia'      (KR, JP)
            Defaults to 'americas'.

    Returns:
        str: The player's PUUID (Universally Unique Identifier) that uniquely
            identifies the account across all Riot games (LoL, Valorant, TFT, etc.).

    Raises:
        KeyError: If the API response doesn't contain a 'puuid' field (e.g., player not found).
        requests.exceptions.RequestException: If the API request fails (network issues, 
            invalid API key, rate limiting, etc.).

    Note:
        This function relies on a globally defined `api_key` variable for authentication.
    '''

    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = f'riot/account/v1/accounts/by-riot-id/{gameName}/{tagLine}'
    print(root_url+endpoint+'?api_key='+api_key)
    response = requests.get(root_url+endpoint+'?api_key='+api_key)

    return response.json()['puuid']
```

```python
def get_gameName_and_tagLine_from_puuid(puuid=None, region='americas'):
    '''Retrieve a player's Riot ID (gameName and tagLine) from their PUUID.

    Performs a reverse lookup on the Riot API to get the human-readable Riot ID
    (e.g., "NOOB HIMA 55#EUNE") associated with a PUUID. This is the inverse
    operation of looking up a PUUID from a Riot ID.

    Args:
        puuid (str, optional): The player's PUUID (Player Universally Unique Identifier).
            This is the 78-character string that uniquely identifies an account across
            all Riot Games titles.
            Example: "zzpN9l1bNdNx4Pm1d7kYz6J8qF3sL2tR8uV5wA0bC1dE2fG3hI4jK5lM6nO7pQ"
            Defaults to None.
        region (str, optional): The Riot API routing region to query.
            Must match the region where the account was originally created.
            Available values:
                - 'americas'  (NA, BR, LAN, LAS)
                - 'europe'    (EUW, EUNE, TR, RU)
                - 'asia'      (KR, JP)
            Defaults to 'americas'.

    Returns:
        dict: A dictionary containing the player's Riot ID components:
            {
                'gameName': str,  # The display name (e.g., "NOOB HIMA 55")
                'tagLine': str    # The tag identifier (e.g., "EUNE")
            }
            These can be combined to form the full Riot ID: "gameName#tagLine"

    Raises:
        KeyError: If the API response doesn't contain 'gameName' or 'tagLine' fields
            (e.g., invalid PUUID or account not found).
        requests.exceptions.RequestException: If the API request fails due to
            network issues, invalid API key, or rate limiting.

    Note:
        - Requires a globally defined `api_key` variable for Riot API authentication.
        - The PUUID lookup is region-specific. Query the wrong region and you'll
          get a 404 even if the account exists elsewhere.

    Example:
        >>> result = get_gameName_and_tagLine_from_puuid("abc123...", region="europe")
        >>> print(f"{result['gameName']}#{result['tagLine']}")
        "NOOB HIMA 55#EUNE"
    '''

    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = 'riot/account/v1/accounts/by-puuid/'

    response = requests.get(root_url+endpoint+puuid+'?api_key='+api_key)

    id = {
        'gameName': response.json()['gameName'],
        'tagLine': response.json()['tagLine']
    }
    
    return id
```

```python
def get_ladder(top=10, region='eun1'):
    '''Get the top players from a region's ranked solo queue ladder.

    Fetches players from Challenger, Grandmaster, and Master tiers as needed.
    To minimize API calls, tiers are only queried if the previous tier doesn't
    contain enough players to satisfy the requested `top` count. Players are
    ranked by League Points (LP) in descending order.

    Args:
        top (int, optional): Number of top players to return.
            Must be a positive integer. Defaults to 10.
            Example: top=50 might fetch Challenger + Grandmaster + some Master players.
        region (str, optional): The regional platform to query.
            Available values:
                'br1'  - Brazil
                'eun1' - Europe Nordic & East
                'euw1' - Europe West
                'jp1'  - Japan
                'kr'   - Korea
                'la1'  - Latin America North
                'la2'  - Latin America South
                'me1'  - Middle East
                'na1'  - North America
                'oc1'  - Oceania
                'ru'   - Russia
                'sg2'  - Singapore
                'tr1'  - Turkey
                'tw2'  - Taiwan
                'vn2'  - Vietnam
            Defaults to 'eun1'.

    Returns:
        pandas.DataFrame: Top `top` players with columns:
            - rank (int): Position on the ladder (1 = highest LP)
            - summonerId (str): Encrypted summoner ID
            - summonerName (str): Player's summoner name
            - leaguePoints (int): Current LP
            - wins (int): Number of wins
            - losses (int): Number of losses
            - veteran (bool): Whether player is a veteran
            - inactive (bool): Whether player is inactive
            - freshBlood (bool): Whether player is new to the tier
            - hotStreak (bool): Whether player is on a hot streak
            - tier (str): Tier name (CHALLENGER, GRANDMASTER, MASTER)

    Raises:
        requests.exceptions.RequestException: If any API call fails.
        KeyError: If API response structure is unexpected.

    Note:
        - Requires a globally defined `api_key` variable.
        - Challenger has ~300 players, Grandmaster ~700, Master can be thousands.
        - API calls are minimized: if top=10, only Challenger is fetched.
        - If top=500, Challenger + Grandmaster are fetched (Master skipped).
        - If top=1500, all three tiers are fetched.
    '''

    root = f'https://{region}.api.riotgames.com/'
    chall = 'lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5'
    gm = 'lol/league/v4/grandmasterleagues/by-queue/RANKED_SOLO_5x5'
    masters = 'lol/league/v4/masterleagues/by-queue/RANKED_SOLO_5x5'
    
    # Initialize DataFrames as empty
    chall_df = pd.DataFrame()
    gm_df = pd.DataFrame()
    masters_df = pd.DataFrame()
    
    # Fetch tiers only as needed
    chall_response = requests.get(root + chall + '?api_key=' + api_key)
    chall_df = pd.DataFrame(chall_response.json()['entries'])
    
    if len(chall_df) < top:
        gm_response = requests.get(root + gm + '?api_key=' + api_key)
        gm_df = pd.DataFrame(gm_response.json()['entries'])
        
        if len(chall_df) + len(gm_df) < top:
            masters_response = requests.get(root + masters + '?api_key=' + api_key)
            masters_df = pd.DataFrame(masters_response.json()['entries'])
    
    # Sort each tier by league points descending
    if not chall_df.empty:
        chall_df = chall_df.sort_values('leaguePoints', ascending=False).reset_index(drop=True)
        chall_df['tier'] = chall_response.json()['tier']
    
    if not gm_df.empty:
        gm_df = gm_df.sort_values('leaguePoints', ascending=False).reset_index(drop=True)
        gm_df['tier'] = gm_response.json()['tier']
    
    if not masters_df.empty:
        masters_df = masters_df.sort_values('leaguePoints', ascending=False).reset_index(drop=True)
        masters_df['tier'] = masters_response.json()['tier']
    
    # Combine all fetched tiers
    ladder = pd.concat([chall_df, gm_df, masters_df], ignore_index=True)
    
    # Create proper ranking (1-based index)
    ladder = ladder.drop(columns=['rank'], errors='ignore')
    ladder = ladder.reset_index(drop=True)
    ladder['rank'] = ladder.index + 1
    
    # Return only the requested top players
    return ladder.head(top)
```

```python
def get_match_history_ids(region='europe', puuid=None, start=0, count=20):
    '''Retrieve a list of match IDs from a player's match history.

    Fetches match IDs for a specific player using their PUUID. Results are 
    paginated using start index and count parameters. The matches are returned 
    in reverse chronological order (most recent first).

    This is the first step in a two-step match data retrieval process:
    1. Get match IDs (this function)
    2. Get detailed match data using those IDs (get_match_data_from_match_id)

    Args:
        region (str, optional): The Riot API routing region to query.
            Determines which regional API server to hit based on where the 
            player's account resides.
            Available values:
                - 'americas'  (NA, BR, LAN, LAS)
                - 'europe'    (EUW, EUNE, TR, RU)
                - 'asia'      (KR, JP)
                - 'sea'       (OCE, SG, PH, TW, VN, TH)
            Defaults to 'europe'.
        puuid (str, optional): The player's PUUID (78-character universally 
            unique identifier). This is the primary player identifier in the 
            current Riot API. Must be provided to fetch match history.
            Defaults to None.
            Example: "SO2yxi2VI3m401bbWQvdwKCG9kNxxlfdOQcRYlTCUpv5rqQZWBrGxOUG90YEg1IV9_h1x6_2bwXraw"
        start (int, optional): The index of the first match to return (0-based).
            Use for pagination to fetch older matches.
            Defaults to 0 (most recent matches).
            Example: start=100 returns matches starting from the 101st most recent.
        count (int, optional): Number of match IDs to return. 
            Valid range: 0-100.
            Defaults to 20.

    Returns:
        list: A list of match ID strings in reverse chronological order.
            Example: ["EUN1_1234567890", "EUN1_1234567891", ...]
            Each match ID is a string that can be used with get_match_data_from_match_id() 
            to retrieve full match details.

    Raises:
        requests.exceptions.RequestException: If the API request fails due to 
            network issues, invalid API key, or rate limiting.

    Note:
        - Requires a globally defined `api_key` variable for authentication.
        - Match IDs are region-specific (e.g., "EUN1_1234567890").
        - Maximum 100 match IDs per request. Use start parameter for pagination.
        - Riot Games retains match data for approximately 2 years.
        - Rate limit: 100 requests per 2 minutes per API key.

    Example:
        # Get the 10 most recent matches
        recent_matches = get_match_history_ids(puuid="abc123...", count=10)
        
        # Get next page (matches 11-30)
        older_matches = get_match_history_ids(puuid="abc123...", start=10, count=20)
    '''

    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = f'lol/match/v5/matches/by-puuid/{puuid}/ids'
    query_params = f'?start={start}&count={count}'

    response = requests.get(root_url + endpoint + query_params + '&api_key=' + api_key)

    return response.json()
```

```python
def get_match_data_from_match_id(region='europe', matchId=None):
    '''Retrieve detailed information about a specific match.

    Fetches comprehensive match data including participant statistics, team 
    compositions, game timeline, and results for a single match. This is 
    the second step in the match data retrieval process.

    The returned data is a large JSON object containing:
    - Match metadata (game version, duration, creation timestamp, queue type)
    - Team-level data (bans, objectives, win status)
    - Player-level data (champion played, KDA, damage, gold, items, runes, etc.)

    Args:
        region (str, optional): The Riot API routing region where the match 
            was played. Must match the region prefix in the matchId.
            Available values:
                - 'americas'  (NA, BR, LAN, LAS)
                - 'europe'    (EUW, EUNE, TR, RU)
                - 'asia'      (KR, JP)
                - 'sea'       (OCE, SG, PH, TW, VN, TH)
            Defaults to 'europe'.
        matchId (str, optional): The unique match identifier string.
            Format: "{platform}_{numeric_id}" (e.g., "EUN1_1234567890").
            Obtained from get_match_history_ids() function.
            Defaults to None.
            Example: "EUN1_1234567890"

    Returns:
        dict: A large JSON object containing complete match data. Key fields include:
            - metadata: matchId, participant PUUIDs, game version
            - info.gameCreation: Timestamp when the game was created
            - info.gameDuration: Length of game in seconds
            - info.queueId: Type of queue (420 = ranked solo, 440 = flex, etc.)
            - info.teams[2]: Team-level data (bans, objectives, win/loss)
            - info.participants[10]: Array of player data including:
                - summonerName, puuid, championName
                - kills, deaths, assists, KDA
                - totalDamageDealtToChampions, goldEarned
                - items, summonerSpells, runes
                - visionScore, wardsPlaced

    Raises:
        requests.exceptions.RequestException: If the API request fails.
        KeyError: If the matchId doesn't exist or the API response structure changes.

    Note:
        - Requires a globally defined `api_key` variable for authentication.
        - Each match detail request counts against your rate limit.
        - Match data can be very large (100KB+ per match) - consider caching.
        - The region parameter MUST match the region where the match was played.
        - Some match data fields may be null for very old matches or certain game modes.

    Example:
        # Get a single match's details
        match_ids = get_match_history_ids(puuid="abc123...", count=1)
        match_data = get_match_data_from_match_id(matchId=match_ids[0])
        
        # Extract player stats
        player = match_data['info']['participants'][0]
        print(f"KDA: {player['kills']}/{player['deaths']}/{player['assists']}")
        print(f"Champion: {player['championName']}")
    '''

    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = f'lol/match/v5/matches/{matchId}'

    response = requests.get(root_url + endpoint + '?api_key=' + api_key)

    return response.json()
```

```python
from dotenv import load_dotenv
import os
import requests
import pandas as pd
import time

load_dotenv()
api_key = os.environ.get('riot_api_key')

# ============================================================================
# RATE LIMITER
# ============================================================================

class RateLimiter:
    '''Rate limiter for Riot Games API.
    
    Enforces Riot's rate limits:
    - 20 requests per 1 second (short-term bucket)
    - 100 requests per 2 minutes (long-term bucket)
    '''
    
    def __init__(self):
        self.short_term_requests = []
        self.long_term_requests = []
        self.short_window = 1
        self.long_window = 120
        self.short_limit = 20
        self.long_limit = 100
    
    def wait_if_needed(self):
        '''Check both rate limits and wait if necessary.'''
        current_time = time.time()
        
        self.short_term_requests = [t for t in self.short_term_requests 
                                     if current_time - t < self.short_window]
        self.long_term_requests = [t for t in self.long_term_requests 
                                    if current_time - t < self.long_window]
        
        if len(self.short_term_requests) >= self.short_limit:
            sleep_time = self.short_term_requests[0] + self.short_window - current_time
            if sleep_time > 0:
                print(f"Short-term rate limit reached. Waiting {sleep_time:.2f}s...")
                time.sleep(sleep_time)
                return self.wait_if_needed()
        
        if len(self.long_term_requests) >= self.long_limit:
            sleep_time = self.long_term_requests[0] + self.long_window - current_time
            if sleep_time > 0:
                print(f"Long-term rate limit reached. Waiting {sleep_time:.2f}s...")
                time.sleep(sleep_time)
                return self.wait_if_needed()
    
    def add_request(self):
        '''Record a new API request.'''
        current_time = time.time()
        self.short_term_requests.append(current_time)
        self.long_term_requests.append(current_time)


# ============================================================================
# EXISTING FUNCTIONS (from your original code)
# ============================================================================

def get_puuid(gameName=None, tagLine=None, region='americas'):
    '''Retrieve a player's PUUID from the Riot Games API.'''
    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = f'riot/account/v1/accounts/by-riot-id/{gameName}/{tagLine}'
    print(root_url + endpoint + '?api_key=' + api_key)
    response = requests.get(root_url + endpoint + '?api_key=' + api_key)
    return response.json()['puuid']


def get_gameName_and_tagLine_from_puuid(puuid=None, region='americas'):
    '''Retrieve a player's Riot ID from their PUUID.'''
    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = 'riot/account/v1/accounts/by-puuid/'
    response = requests.get(root_url + endpoint + puuid + '?api_key=' + api_key)
    id = {
        'gameName': response.json()['gameName'],
        'tagLine': response.json()['tagLine']
    }
    return id


def get_match_history_ids(region='europe', puuid=None, start=0, count=20):
    '''Retrieve a list of match IDs from a player's match history.'''
    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = f'lol/match/v5/matches/by-puuid/{puuid}/ids'
    query_params = f'?start={start}&count={count}'
    response = requests.get(root_url + endpoint + query_params + '&api_key=' + api_key)
    return response.json()


def get_match_data_from_match_id(region='europe', matchId=None):
    '''Retrieve detailed information about a specific match.'''
    root_url = f'https://{region}.api.riotgames.com/'
    endpoint = f'lol/match/v5/matches/{matchId}'
    response = requests.get(root_url + endpoint + '?api_key=' + api_key)
    return response.json()


# ============================================================================
# NEW FUNCTION - THE MISSING PIECE
# ============================================================================

# def get_player_ids_from_match(matchId=None, region='europe'):
#     '''Extract all player PUUIDs from a single match.
    
#     Retrieves the 10 participant PUUIDs from a given match.
    
#     Args:
#         matchId (str): The unique match identifier string.
#         region (str): The Riot API routing region. Defaults to 'europe'.
    
#     Returns:
#         list: A list of 10 PUUID strings for each player in the match.
#     '''
#     match_data = get_match_data_from_match_id(region=region, matchId=matchId)
#     puuid_list = []
#     for participant in match_data['info']['participants']:
#         puuid_list.append(participant['puuid'])
#     return puuid_list


def get_player_ids_from_match(matchId=None, region='europe', solo_queue_only=True):
    '''Extract all player PUUIDs from a single match.
    
    Args:
        matchId (str): The unique match identifier string.
        region (str): The Riot API routing region. Defaults to 'europe'.
        solo_queue_only (bool): If True, only return players from Ranked Solo/Duo games.
            Defaults to True.
    
    Returns:
        list: A list of 10 PUUID strings, or empty list if not a solo queue match.
    '''
    match_data = get_match_data_from_match_id(region=region, matchId=matchId)
    
    # Filter for Ranked Solo/Duo only (queueId = 420)
    if solo_queue_only and match_data['info']['queueId'] != 420:
        print(f"  Skipping - Not a Ranked Solo/Duo game (queueId: {match_data['info']['queueId']})")
        return []
    
    puuid_list = []
    for participant in match_data['info']['participants']:
        puuid_list.append(participant['puuid'])
    return puuid_list


# ============================================================================
# PLAYER COLLECTION FUNCTIONS
# ============================================================================


def get_player_ids_from_matches(match_ids=None, region='europe', max_players=100, rate_limiter=None):
    '''Extract up to N unique player PUUIDs from a list of match IDs.'''
    if not match_ids:
        raise ValueError("match_ids list cannot be empty")
    
    if rate_limiter is None:
        rate_limiter = RateLimiter()
    
    unique_puuids = []
    matches_processed = 0
    match_ids_used = []
    
    for i, match_id in enumerate(match_ids):
        if len(unique_puuids) >= max_players:    # ← Now dynamic
            break
        
        try:
            rate_limiter.wait_if_needed()
            
            print(f"Processing match {i+1}/{len(match_ids)}: {match_id} "
                  f"(Collected: {len(unique_puuids)}/{max_players} players)")
            
            players = get_player_ids_from_match(matchId=match_id, region=region)
            rate_limiter.add_request()
            matches_processed += 1
            
            for puuid in players:
                if puuid not in unique_puuids and len(unique_puuids) < max_players:  # ← Dynamic
                    unique_puuids.append(puuid)
                    if match_id not in match_ids_used:
                        match_ids_used.append(match_id)
        
        except Exception as e:
            print(f"Error processing match {match_id}: {e}")
            rate_limiter.add_request()
            continue
    
    print(f"\nCompleted! Processed {matches_processed} matches, "
          f"found {len(unique_puuids)} unique players")
    
    return {
        'puuids': unique_puuids,
        'count': len(unique_puuids),
        'matches_processed': matches_processed,
        'match_ids_used': match_ids_used
    }



def get_player_ids_from_player(puuid=None, region='europe', max_matches=20, max_players=100, rate_limiter=None):
    '''Get N unique player PUUIDs by snowball sampling from one starting player.'''
    if not puuid:
        raise ValueError("Starting PUUID is required")
    
    if rate_limiter is None:
        rate_limiter = RateLimiter()
    
    print(f"Fetching {max_matches} recent matches for starting player...")
    rate_limiter.wait_if_needed()
    match_ids = get_match_history_ids(
        region=region, 
        puuid=puuid, 
        start=0, 
        count=max_matches
    )
    rate_limiter.add_request()
    
    print(f"Found {len(match_ids)} matches to process")
    
    result = get_player_ids_from_matches(
        match_ids=match_ids, 
        region=region, 
        max_players=max_players,    # ← Pass through
        rate_limiter=rate_limiter
    )
    
    if puuid in result['puuids']:
        result['puuids'].remove(puuid)
        result['count'] -= 1
    
    result['start_puuid'] = puuid
    return result


# ============================================================================
# EXAMPLE USAGE
# ============================================================================

if __name__ == "__main__":
    my_puuid = get_puuid(gameName='NOOB HIMA 55', tagLine='eune', region='europe')
    
    limiter = RateLimiter()
    start_time = time.time()
    
    # result = get_100_player_ids_from_player(
    #     puuid=my_puuid, 
    #     max_matches=20, 
    #     rate_limiter=limiter
    # )




    # Get 50 players
    # result = get_player_ids_from_player(puuid=my_puuid, max_players=50)
    
    # # Get 500 players (will take ~2 minutes)
    # result = get_player_ids_from_player(puuid=my_puuid, max_matches=80, max_players=500)
    
    # # Get 1000 players (will take ~5-7 minutes due to rate limiting)
    # result = get_player_ids_from_player(puuid=my_puuid, max_matches=100, max_players=1000)
    


    result = get_player_ids_from_player(puuid=my_puuid, max_matches=80, max_players=200)

    end_time = time.time()
    
    print(f"\n{'='*50}")
    print(f"FINAL RESULTS:")
    print(f"Total time: {end_time - start_time:.2f} seconds")
    print(f"Unique players collected: {result['count']}")
    print(f"Matches processed: {result['matches_processed']}")
    print(f"Match IDs used: {result['match_ids_used']}")
```

```python
import os
import pandas as pd
from datetime import datetime

def save_player_ids_to_csv(result_dict, filename=None):
    '''Save collected PUUIDs to a CSV file with date in filename.
    
    Args:
        result_dict (dict): The dictionary returned by get_player_ids_from_player()
            or get_player_ids_from_matches(). Must contain 'puuids' key.
        filename (str, optional): Base name for the CSV file. Date will be appended.
            If None, uses 'player_puuids' as base. Defaults to None.
    
    Returns:
        str: The full filepath where the CSV was saved.
    '''
    current_dir = os.getcwd()
    data_dir = os.path.abspath(os.path.join(current_dir, '..', 'data', 'data collected'))
    os.makedirs(data_dir, exist_ok=True)
    
    # Always add timestamp
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    
    if filename is None:
        filename = f'player_puuids_{timestamp}.csv'
    else:
        # Insert timestamp before the .csv extension
        base_name = filename.replace('.csv', '')
        filename = f'{base_name}_{timestamp}.csv'
    
    filepath = os.path.join(data_dir, filename)
    
    df = pd.DataFrame({'puuid': result_dict['puuids']})
    df.to_csv(filepath, index=False)
    
    print(f"Saved {len(result_dict['puuids'])} PUUIDs to: {filepath}")

    df.to_csv(filepath, index=False)
    
    return filepath
```
